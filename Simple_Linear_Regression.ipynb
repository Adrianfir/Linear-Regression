# Simple Linear Regression

## Import Packages

```python
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow import keras

import matplotlib.pyplot as plt
```

## Simple Data

```python
a = 2
b = 0.9

x = np.linspace(0,5,250)
noise  = np.random.randn(x.shape[0])*1.8

y = a*x + b +noise

plt.plot(x,y,'o',label = 'Data Points')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Dataset')
plt.legend(loc="best")
plt.show()
```

## Define Model

```python
class Linear_regression_model:
    def __init__ (self):
        self.weight = tf.Variable(2.0)
        self.bias = tf.Variable(1.0)
    
    def __call__(self,x):
        return self.weight*x + self.bias
```

## Define Loss Function

```python
def loss(y,y_pred):
    return tf.reduce_mean(tf.square(y-y_pred))
```

## Train

```python
def train(model, x, y, lr):
    
    with tf.GradientTape() as tp:
        current_loss = loss(y,model(x))    ## Define Objective Function of the problem here for training
        
    lr_weight, lr_bias = tp.gradient(current_loss,[model.weight, model.bias])
    
    model.weight.assign_sub(lr*lr_weight)
    model.bias.assign_sub(lr*lr_bias)
```

##  Main Program

```python
model = Linear_regression_model()

epochs = 1000
loss_val = np.zeros(epochs)

for i in range(epochs):
    loss_val[i] = loss(y,model(x)).numpy()
    train(model, x, y, 0.01)
    print("Epoch Number = ", i, " Loss= ", loss_val[i])
    
plt.plot(loss_val, label='Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
plt.legend(loc='best')
```

## Plot The Result

```python
plt.plot(x,y,'o',label = 'Data Points')
plt.plot(x,model.weight*x + model.bias, label = 'Linear_Regression Result')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Dataset')
plt.legend(loc="best")
plt.show()

print('Weight: ',model.weight.numpy(), '----- Bias: ', model.bias.numpy())
```
